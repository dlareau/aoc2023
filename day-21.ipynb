```python
class WrapGrid(np.ndarray):
    def __new__(cls, grid_str, conv_func=str):
        grid = [[conv_func(p) for p in row] for row in grid_str.split("\n")]
        obj = np.asarray(grid).view(cls)
        return obj
    
    def sides(self, p, both=False):
        row, col = p
        side_list = []
        if(row != 0):
            side_list.append(((row-1, col), (0, 0)))
        else:
            side_list.append(((len(self)-1, col), (-1, 0)))

        if(col != 0):
            side_list.append(((row, col-1), (0, 0)))
        else:
            side_list.append(((row, len(self[0])-1), (0, -1)))

        if(row != len(self) - 1):
            side_list.append(((row+1, col), (0, 0)))
        else:
            side_list.append(((0, col), (1, 0)))

        if(col != len(self[0]) - 1):
            side_list.append(((row, col+1), (0, 0)))
        else:
            side_list.append(((row, 0), (0, 1)))

        if both:
            vals_side_list = [self[point[0][0]][point[0][1]] for point in side_list]
            return list(zip(side_list, vals_side_list))
        return side_list
```

```python
def add_tuples(a, b):
    return ((a[0] + b[0]), (a[1] + b[1]))

def get_locations(grid, start, num_steps):
    front = collections.defaultdict(int)
    front[start] = 1
    for i in range(num_steps):
        new_front = collections.defaultdict(int)
        places_front = collections.defaultdict(set)
        for point in front:
            new_points = grid.sides(point, both=True)
            for new_point in new_points:
                location, val = new_point
                if val == "#":
                    continue
                p_loc, grid_loc = location
                if grid_loc in places_front[p_loc]:
                    
                if grid_loc != (0, 0):
                    old_locations = [add_tuples(o, grid_loc) for o in old_locations]
                new_front[p_loc].update(old_locations)
        front = new_front
    return front
```
