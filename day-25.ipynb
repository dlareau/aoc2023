```python
from aocd.models import Puzzle
import networkx as nx
puzzle = Puzzle(year=2023, day=25)
data = puzzle.input_data
%run helper.ipynb
# data= """jqt: rhn xhk nvd
# rsh: frs pzl lsr
# xhk: hfx
# cmg: qnr nvd lhk bvb
# rhn: xhk bvb hfx
# bvb: xhk hfx
# pzl: lsr hfx nvd
# qnr: nvd
# ntq: jqt hfx bvb xhk
# nvd: lhk
# lsr: lhk
# rzs: qnr cmg lsr rsh
# frs: qnr lhk lsr"""
```

```python
G = nx.Graph()
lines = data.split("\n")
for line in lines:
    src, dests = line.split(": ")
    for dst in dests.split(" "):
        G.add_edge(src, dst, label=f"{src}/{dst}")
```

```python
len(G.edges())
```

```python
pos = nx.spring_layout(G)
nx.draw(G, pos)
nx.draw_networkx_edge_labels(G, pos,edge_labels=nx.get_edge_attributes(G,'label'))
```

```python
# pcc/htj | dlk/pjj | htb/bbg
G.remove_edge("pcc", "htj")
G.remove_edge("dlk", "pjj")
G.remove_edge("htb", "bbg")
```
